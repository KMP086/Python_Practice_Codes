#Excercise Guessing Game



1.   #generate a number 1~10
2.   #input from user?
3.   #check that input is a number 1~10
4.   #check if number is the right guess. Otherwise
5.   #ask again



```python
from random import randint
#generate a number 1~10
answer = randint(1,10)


#check that input is a number 1~10
while True:
  try:
#input from user?

    guess = input('guess a number 1-10: ')

    if int(guess) > 1 and int(guess) < 11:
      print('all good')
      break
  except ValueError:
    print('please enter a number')
    continue

#check if number is the right guess. Otherwise

#ask again
```

```python
from random import randint
import sys
#generate a number 1~10
answer = randint(int(sys.argv[1]),int(sys.argv[2]))

#check that input is a number 1~10
while True:
  try:
#input from user?

    guess = int(input('guess a number 1-10: '))

    if 0 < guess < 11:
      if guess == answer:
        print('you are a genius!')
      break
    else:
      print('hey bozo, I said  1~10')
  except ValueError:
    print('please enter a number')
    continue

#check if number is the right guess. Otherwise

#ask again
```

#Useful Modules

```python
#Specialized Data Type

from collections import Counter, defaultdict, OrderedDict
li = [1,2,3,4,5,6,7,7]
# The display is in dictionary form
print(Counter(li))

sentence = 'lah blah blah thingking about python'
print(Counter(sentence))

#Counter is best in counting duplicate items
```

```python
dictionary = {'a':1, 'b':2}
print(dictionary['a'])
#error print(dictionary['c'])

#to make default
dictionary2 = defaultdict(int,{'a':1, 'b':2})
#Still error print(dictionary['c'])

dictionary3 = defaultdict(lambda: 'does not exit', {'a':1, 'b':2})
print(dictionary3['c'])
```

```python
d = OrderedDict()
d['a'] =1
d['b'] =2

d2 = OrderedDict()
d2['b'] = 2
d2['a'] = 1

print(d2==d)
```
