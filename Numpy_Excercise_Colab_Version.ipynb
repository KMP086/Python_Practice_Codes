##Dot product example(nut butter sales)

```python
np.random.seed(0)
sales_amounts = np.random.randint(20,size=(5, 3))
sales_amounts
```

```python
#create weekly_sales dataframe
weekly_sales = pd.DataFrame(sales_amounts,
                            index = ["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns=["Almond butter",
                                     "Peanut butter",
                                     "Cashew butter"])
weekly_sales
```

```python
#Create prices array
prices = np.array([10, 8, 12])
prices
```

```python
prices.shape #this will result to error #reshape this one by adding 1
```

```python
#Create butter_prices Dataframe
butter_prices = pd.DataFrame(prices.reshape(1,3),
                             index=["Price"],
                             columns=["Almond butter",
                                     "Peanut butter",
                                     "Cashew butter"])
butter_prices
```

```python
prices.reshape(1,3).shape
```

```python
sales_amounts.shape
```

```python
#total_sales = prices.dot(sales_amounts) error
#total_sales
```

```python
#shape aren't aligned, lets transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
#Create daily sales
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
#dialy sales is on row  need to be in column
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
